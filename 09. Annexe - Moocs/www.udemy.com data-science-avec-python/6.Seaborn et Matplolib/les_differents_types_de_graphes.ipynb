```python
import matplotlib.pyplot as plt
import numpy 
import pandas
# Réaliser des stats sur un fichier de ventes de black friday .
black_friday = pandas.read_csv("BlackFriday.csv")
```

```python
# Combien de femmes et combiens d'hommes ont fait des achats ?
# Faire un diagramme en barre 
data = black_friday['Gender'].value_counts()
type(data)
```

```python
plt.rcParams["figure.figsize"] = [ 16,9 ]
plt.bar(data.index,data,color=["blue","red"])
plt.xlabel("sexe")
plt.ylabel("NOmbre")
plt.title("Répartition Femme et Homme :",y = 1.05)
plt.draw()
plt.savefig("mon_graphique.png",bbox_inches="tight")
```

```python
# Créer un box plot pour savoir la distribution des achats.
plt.rcParams["figure.figsize"] = [ 16,9 ]
plt.boxplot(black_friday["Purchase"])
plt.xticks([1],["Ventes"])
plt.draw()
```

```python
# Créer un camembert piechart des catégories d'ages

data = black_friday["Age"].value_counts()
data
```

```python
# Créer un camembert pie pour savoir la distribution des achats.
plt.rcParams["figure.figsize"] = [ 16,9 ]
# On donne a pie les valeurs des intervalles d'ages
plt.pie(data.values,labels=data.index,autopct='%1.1f%%')
plt.xticks([1],["Ventes"])
plt.draw()
```
