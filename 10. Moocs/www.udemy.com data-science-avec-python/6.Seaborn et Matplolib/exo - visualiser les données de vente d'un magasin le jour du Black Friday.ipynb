```python
import matplotlib.pyplot as plt
import numpy 
import pandas
import seaborn as sns

# Exercice de fin dechapitre sur seaborn
tips = sns.load_dataset('tips')
tips.columns
```

```python
tips
```

```python
# 1. Tracer un graphique représentant la variable total_bill, séparée par sexe
sns.barplot(y="total_bill",x="sex",data = tips)
plt.draw()
```

```python
# 2. Etudier la distribution de la variable total_bill à l'aide de la fonction sns.distplot()
# Je prends toutes les valeurs de la colonne total_bill puis les mets dans une liste python
toutes_mes_vars = tips['total_bill'].to_list()
# Du coup, ensuite, je peux étudier la distribution.
sns.distplot(toutes_mes_vars)

# CORRECTION densité de probabilité d'une variable aléatoire ( C'était plus ssimple que ce  que j'ai fait.)
sns.distplot(tips['total_bill'])
```

```python
# 3. Utiliser la function jointplot de seaborn pour tracer la variable total_bill X en fonction de la variable tip (Y)
# C'est un graphique bi-varié : étude d'une relation entre 2 variables . et univarié sur les barres
g = sns.jointplot(x="total_bill", y="tip", data=tips)
g
```

```python
# Correction : Elle fait une regression linéaire en plus 
g = sns.jointplot(x="total_bill", y="tip", data=tips,kind = "reg")
g
```

```python
# 4. Utiliser la function pairplot de seaborn sur l'ensemble du dataframe 'tips' pour étudier les relations sur l'ensemble des variables

g = sns.pairplot(data=tips)
g
```

```python
# 5. Tracer le boxplot de la variable total_bill, en créant un boxplot par jour (variable day) et en utilisant la variable sex dans l'option 'hue'
sns.boxplot(data = tips, x = "total_bill",y="day" ,hue = "sex")
# Correction : "le jeudi, les hommes semblent payer plus que les femmes.""
```
