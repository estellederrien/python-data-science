```python
# EXAMEN  : Etudier le marché Android Store

# 1. Quels sont les noms, les genres et les catégories des applications les plus installées sur le google Play ?

# Réponse : On veut déjà connaitres les noms des apps les plus téléchargées.
# a) On commence par lire le fichier
import pandas as pd
google_app = pd.read_table("googleplaystore.csv",sep=",")
google_app

```

```python
# On rêgle l'affichage du tableau pour bien voir toutes les colonnes dans jupyterlab
# On affiche que 10 lignes
pd.set_option('display.max_rows', 20)
pd.set_option('display.max_columns', 500)
# Largeur de l'écran
pd.set_option('display.width', 400)
```

```python
# b) Pour trier par nombre d'installs, il faut oter le caractère + et , de chaque ligne de la colonne "Installs", puis caster en INT :

# Remplacer des var  qualitatives en quantitatives 

google_app["Installs"] = google_app["Installs"].apply(lambda x: x.replace(",",""))
google_app["Installs"] = google_app["Installs"].apply(lambda x: x.replace("+",""))
google_app["Installs"] = google_app["Installs"].apply(lambda x: x.replace("Free","0"))

# Caster la colonne Installs en INT :
google_app['Installs'] = google_app['Installs'].astype(int)    

# Montrer les 5 premières apps les plus installées
google_app.sort_values(by='Installs', ascending=False)
```

```python
# Réponse 2 : Connaitre quels est le genre des apps les plus téléchargées

# On groupe par genre
Genres = google_app.groupby(["Genres"])

# Avec describe, on voit que le genre ayant le plus d'apps téléchargées est Action
Genres.describe(include ="all")
```

```python
# Réponse 3 : On veut connaitre les catégories des apps les plus téléchargées.

# On groupe par category
Categories = google_app.groupby(["Category"])

#On peut afficher chaque lignes par categories
#for grp in Categories:
    #print (grp)

# Avec describe, on voit que la catégorie ayant le plus d'apps téléchargées est Tools, (outils)
Categories.describe(include ="all")
```

```python
# on doit d'abord nettoyer la colonne   	Reviews car il n'y a pas que du INT , il y a des caractéres numériques, du coup, les stats sont faussées :

# Remplacer des var  qualitatives en quantitatives ( les M veulent dire Millions, les K veulent dire Milliers, mais pour faire les stats, il ne les comprends pas, on passe tout en INT !)
google_app['Reviews'] = google_app['Reviews'].replace({'K': '*1e3', 'M': '*1e6'}, regex=True).map(pd.eval).astype(int)

google_app.sort_values(by='Reviews', ascending=False)
```

```python
#Avec column.max()
column = google_app["Reviews"]
max_value = column.max()
print(max_value)
```
