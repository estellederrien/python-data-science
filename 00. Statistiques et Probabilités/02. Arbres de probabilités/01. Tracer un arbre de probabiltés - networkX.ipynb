# Dessiner un arbre de probabilité 
avec python networkX par Estelle Derrien

```python
# Import des libs
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
```

```python


# paramètre des arcs
edges = [['A', 'B'], ['B', 'C'], ['B', 'D']]

# Graph orienté
G = nx.DiGraph() 

G.add_edges_from(edges)

position = nx.spring_layout(G)
plt.figure()
nx.draw(
    G,  edge_color='black', width=1, linewidths=1,
    node_size=500, node_color='pink', alpha=0.9,
    labels={node: node for node in G.nodes()},pos=position
)

# Valuation des arcs
nx.draw_networkx_edge_labels(
    G, position,
    edge_labels={('A', 'B'): '1/3', 
                 ('B', 'C'): '1/4', 
                 ('B', 'D'): '1/4'},
    font_color='red'
)
plt.axis('off')
plt.show()
```

On voit que l'arbre est correct mais pas bien positionné de gauche à droite ( by Estelle)

```python



# paramètre des arcs
arcs = [['A', 'B'], ['B', 'C'], ['B', 'D']]

# Graph orienté
G = nx.DiGraph() 

G.add_edges_from(arcs)

position = nx.spring_layout(G)

plt.figure()

# Position des nodes
position["A"] = np.array([-3, 0])  # On spécifie la position du node de cette façon
position["B"] = np.array([0, 0])  # On spécifie la position du node de cette façon
position["C"] = np.array([3, 4])  # On spécifie la position du node de cette façon
position["D"] = np.array([3, -4])  # On spécifie la position du node de cette façon

nx.draw(
    G,  edge_color='black', width=1, linewidths=1,
    node_size=500, node_color='pink', alpha=0.9,
    labels={node: node for node in G.nodes()},pos=position
)

# Valuation des arcs
nx.draw_networkx_edge_labels(
    G, position,
    edge_labels={('A', 'B'): '1/3', 
                 ('B', 'C'): '1/4', 
                 ('B', 'D'): '1/4'},
    font_color='red'
)

plt.axis('off')
plt.show()
```
