# Calculer une intégrale et l'afficher

- Rôle : Calculer l'aire sous une courbe entre deux points A et B.
- Intérêt : Calculer l'aire d'un champs pas carré, établir des statistiques etc ...

A quoi ça ressemble : 

<img src="img/sphx_glr_integral_001.webp">

```python
# On importe les librairies
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
import scipy.integrate as spi  


# On crée une fonction
def func(x):
    return (x - 3) * (x - 5) * (x - 7) + 85

# On crée les points A et B
a, b = 2, 9  


# On calcule l'aire sous la courbe entre A et B  avec Scipy 
integration, error = spi.quadrature(func, a, b)  
print('l\'intégrale de la fonction donnée est ', integration, "avec un taux d\'erreur de ", error)  


```

```python
# Maintenant, on va dessiner la fonction et le look de l'aire sous la courbe.

x = np.linspace(0, 10)
y = func(x)

fig, ax = plt.subplots()
ax.plot(x, y, 'r', linewidth=2)
ax.set_ylim(bottom=0)

# Make the shaded region
ix = np.linspace(a, b)
iy = func(ix)
verts = [(a, 0), *zip(ix, iy), (b, 0)]
poly = Polygon(verts, facecolor='0.9', edgecolor='0.5')
ax.add_patch(poly)

ax.text(0.5 * (a + b), 30, r"$\int_a^b f(x)\mathrm{d}x$",
        horizontalalignment='center', fontsize=20)

fig.text(0.9, 0.05, '$x$')
fig.text(0.1, 0.9, '$y$')

ax.spines[['top', 'right']].set_visible(False)
ax.set_xticks([a, b], labels=['$a$', '$b$'])
ax.set_yticks([])

plt.show()
```
